```python
class Product:
    def __init__(self, name, stock):
        self.name = name
        self.stock = stock

    def move(self, from_location, to_location, quantity):
        if from_location == to_location or quantity <= 0 or self.stock[from_location] < quantity:
            return False
        self.stock[from_location] -= quantity
        self.stock[to_location] = self.stock.get(to_location, 0) + quantity
        return True

# Example usage
product_stock = {
    "Warehouse A": 100,
    "Warehouse B": 50
}

product = Product("Product A", product_stock)
print(f"Initial stock of {product.name}: {product.stock}")

moved_successfully = product.move("Warehouse A", "Warehouse B", 50)
if moved_successfully:
    print("Stock moved successfully!")
else:
    print("Failed to move stock due to insufficient stock or invalid locations.")

print(f"Stock of {product.name} at Warehouse A: {product.stock.get('Warehouse A', 0)}")
print(f"Stock of {product.name} at Warehouse B: {product.stock.get('Warehouse B', 0)}")
```

```python
class Movement:
    def __init__(self,product,quantity,date):
        self.product = product
        self.quantity = quantity
        self.date = date
          
        
    def movements_by_product(product, movement_list):
        return [movement for movement in movement_list if movement.product==product]
    
    
movement_list = [
        Movement("wax heater", 20, "2022-08-01"),
        Movement("wax", 10, "2022-08-01"),
        Movement("wax heater", -5,  "2022-08-01"),
        Movement("wax strip", 8, "2022-08-01"),
    ]
    
product_names = ["wax heater","wax","wax strip"]

for product_name in product_names:
    product_movements = Movement.movements_by_product(product_names, movement_list)
    print(f"Movement history for {product_name}:")
    for movement in product_movements:
        print(f"Quantity: {movement.quantity}, Date: {movement.date}")
    print()    
    

    
```

```python
class Product:
    def __init__(self, name, location, stock):
        self.name = name
        self.location = location
        self.stock = stock


product_list = [
    Product("Product A", "Warehouse A", 100),
    Product("Product B", "Warehouse A", 75),
    Product("Product C", "Warehouse B", 50),
    Product("Product D", "Warehouse B", 30),
    Product("Product E", "Warehouse C", 20),
]

location_products = {}

for product in product_list:
    if product.location in location_products:
        location_products[product.location].append(product)
    else:
        location_products[product.location] = [product]

for location, products in location_products.items():
    print(f"Location: {location}")
    for product in products:
        print(f"Product: {product.name}, Stock: {product.stock}")
    print()
```
